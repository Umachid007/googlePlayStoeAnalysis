```python
playstore['Installs'] = playstore['Installs'].str.replace('+', ' ')
playstore['Installs'] = playstore['Installs'].str.replace(',', '')
playstore["Installs"]=playstore["Installs"].str.strip()
playstore["Installs"]=playstore['Installs'].astype(int)
```

```python
playstore['Rating'].fillna(playstore['Rating'].mean(), inplace=True)
```

```python
#Counting Rating for each category
playstore.groupby('Category')['Rating'].value_counts().sort_values(ascending=False)
```

```python
playstore['Android Ver'] = playstore['Android Ver'].str.replace('and up', ' ')
```

```python
playstore['Rating']=round(playstore['Rating'],2)
```

```python
round(playstore.groupby('Category')['Rating'].mean(),2)
```

```python
playstore['Size'] = playstore['Size'].str.replace('M', '')
playstore['Size'] = playstore['Size'].str.replace('k', '')
playstore['Size']=playstore['Size'].str.replace('Varies with device','24')
playstore['Size']=playstore['Size'].str.strip()
playstore["Size"]=playstore['Size'].astype(float)
playstore.info()
```

```python
playstore['Price'] = playstore['Price'].str.replace('$', '')
playstore['Price']=playstore['Price'].str.strip()
playstore["Price"]=playstore['Price'].astype(float)
```

Convert correlation table to heat map

```python
result = playstoreCorre.corr()
result = result [result < 1] ## To avoid getting correlation of variable with themselves i.e value of 1 
sns.heatmap(result ,annot=True,cmap='Blues')
```
